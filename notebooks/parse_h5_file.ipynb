```python
import h5py
import pandas as pd
import numpy as np

def list_h5_metadata_py(h5_file, num_examples=2):
    with h5py.File(h5_file, 'r') as f:
        # List all groups and datasets in the file
        print("File Structure:")
        f.visit(print)
        
        # Function to recursively print group names
        def print_attrs(name, obj):
            if isinstance(obj, h5py.Dataset):
                print("\nDataset:", name)
                data = obj[()]
                
                # Print column names and a few example rows
                if isinstance(data, np.ndarray) and data.dtype.names:
                    print("Column names:", data.dtype.names)
                    df = pd.DataFrame(data)
                    print("Example rows:")
                    print(df.head(num_examples))
                elif isinstance(data, np.ndarray):
                    print("Data shape:", data.shape)
                    print("Example rows:")
                    print(data[:num_examples])
                else:
                    print("Data:")
                    print(data)

        # Print metadata and example rows
        f.visititems(print_attrs)

# Example usage
h5_file = "DESeq2_filtered_exon_intron_abnormal_conditions_all_reads.h5"
list_h5_metadata_py(h5_file)
```
